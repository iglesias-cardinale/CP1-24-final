```python
import distance as ds
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, lat, lon, alt, alt84, speed, direction, distance, sigma_horz_meters, sigma_alt_meters, satellites = np.loadtxt(
    '/workspaces/CP1-24-midterm/iglesias-cardinale/data/ic001_gps_circle_run001.csv' ,
    skiprows=1,
    delimiter = ',',
    unpack=True)
```

```python
#Convert latitude and longitude from degrees to radians

lat_rad = lat*np.pi/180
lon_rad = lon*np.pi/180
```

```python
#Calculate the x, y and z positions

x,y,z = ds.gps_wgs84(lat_rad, lon_rad, alt84)
```

```python
#Plot the data

plt.plot(x,y)
```

```python
#Find distance between start and end points

p1 = [lat_rad[0], lon_rad[0], alt84[0]]
p2 = [lat_rad[-200], lon_rad[-200], alt84[-200]]

dist = [ds.gps_distance_wgs84(p1, [lat_rad[i], lon_rad[i], alt84[i]]) for i in range(len(lat_rad))]

plt.plot(t, dist)
```
